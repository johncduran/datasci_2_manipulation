```python
dataset_path = 'https://raw.githubusercontent.com/johncduran/datasci_2_manipulation/main/datasets/healthcare_data_cleaning.csv'
```

```python

start_time = time.time()
polars_df = pl.read_csv(dataset_path)
polars_filtered = polars_df.filter(polars_df['Gender'] == 'Male')
polars_average_age = polars_filtered['Patient Age'].mean()
end_time = time.time()
polars_time = end_time - start_time
print(f'polars loading time {polars_time:.4f} seconds')
```
